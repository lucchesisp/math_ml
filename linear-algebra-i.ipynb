# System of Linear Equations

Here is an example code showing how to view a system of linear equations.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Minutes 0 -> 40 with 1000 dots
t = np.linspace(0, 40, 1000)
```

```python
# Distance travelled by robber: d = 2.5t
dr = 2.5 * t

# Distance travelled by sheriff: d = 3(t - 5)
ds = 3 * (t-5)
```

```python
# Plot
_, ax = plt.subplots()
plt.title("A Bank Robber Caught")
plt.xlabel("Time (in minutes)")
plt.ylabel("Distance (in km)")
ax.set_xlim([0, 40])
ax.set_ylim([0, 100])
ax.plot(t, dr, c='green')
ax.plot(t, ds, c='brown')
plt.axvline(x=30, color='purple', linestyle='--')
plt.axhline(y=75, color='purple', linestyle='--')
```

# **Hadamard Product**

$
\mathbf{C} = \mathbf{A} \circ \mathbf{B}
$

##### Description:
* The **Hadamard product** (also called the element-wise product) of two matrices $ \mathbf{A} $ and $ \mathbf{B} $ is calculated by multiplying corresponding elements of the matrices.
* This operation requires that both matrices $ \mathbf{A} $ and $ \mathbf{B} $ have the same dimensions.

##### Formula Details:
- For matrices $ \mathbf{A} $ and $ \mathbf{B} $ of size $ m \times n $:
  $
  C_{ij} = A_{ij} \cdot B_{ij}, \quad \forall i \in [1, m], \, j \in [1, n]
  $
  where:
  - $ C_{ij} $ is the element of the resulting matrix $ \mathbf{C} $ at row $ i $ and column $ j $.
  - $ A_{ij} $ and $ B_{ij} $ are the elements of $ \mathbf{A} $ and $ \mathbf{B} $, respectively, at the same position.

##### Practical Interpretation:
* The Hadamard product performs element-wise multiplication of two matrices.
* It is commonly used in machine learning, deep learning, and linear algebra when combining two matrices without altering their structure.

##### Example:
If:
$
\mathbf{A} =
\begin{bmatrix}
1 & 2 \\
3 & 4
\end{bmatrix}, \quad
\mathbf{B} =
\begin{bmatrix}
5 & 6 \\
7 & 8
\end{bmatrix}
$

Then the Hadamard product $ \mathbf{C} = \mathbf{A} \circ \mathbf{B} $ is:
$
\mathbf{C} =
\begin{bmatrix}
1 \cdot 5 & 2 \cdot 6 \\
3 \cdot 7 & 4 \cdot 8
\end{bmatrix} =
\begin{bmatrix}
5 & 12 \\
21 & 32
\end{bmatrix}
$

# L Norm of a Vector
$
\|\mathbf{v}\|_2 = \sqrt{\sum_{i=1}^n v_i^2}
$

##### Description:
* The L norm of a vector $v$ is calculated as the square root of the sum of the squares of its elements.
* The summation indicates that we are summing the squares of all $n$ elements of the vector.
* $v_i$ represents the value of the $i-th$ element of the vector.

##### Practical Interpretation:
* The L norm measures the "distance" of the vector $v$ from the origin in Euclidean space.
* It is the most commonly used norm, also called the Euclidean norm.

# **Frobenius Norm**

$
\|\mathbf{X}\|_F = \sqrt{\sum_{i,j} X_{ij}^2}
$

- **Description:**
  The **Frobenius norm** of a matrix $\mathbf{X}$ is calculated as the square root of the sum of the squares of all its elements.

- **Formula Details:**
  - The double summation $\sum_{i,j}$ indicates that all elements $X_{ij}$ of the matrix are considered, where:
    - $i$ indexes the rows.
    - $j$ indexes the columns.
  - $X_{ij}$ is the value of the element in the $i-th$ row and $j-th$ column of $\mathbf{X}$.


**Interpretation:**
  - The Frobenius norm measures the overall "magnitude" or "size" of the matrix.
  - It treats all elements of the matrix as if they were part of a flattened vector and calculates the **Euclidean norm** of this vector.

# **Matrix Multiplication**

$
C_{i,k} = \sum_{j} A_{i,j} B_{j,k}
$

##### Description:
* The **matrix multiplication** computes a resulting matrix $ \mathbf{C} $, where each element $ C_{i,k} $ is calculated as the sum of the products of corresponding elements from row $ i $ of matrix $ \mathbf{A} $ and column $ k $ of matrix $ \mathbf{B} $.

##### Formula Details:
- For matrices $ \mathbf{A} $ of size $ m \times n $ and $ \mathbf{B} $ of size $ n \times p $:
  - The resulting matrix $ \mathbf{C} $ will have dimensions $ m \times p $.
  - Each element $ C_{i,k} $ is calculated as:
    $
    C_{i,k} = \sum_{j=1}^n A_{i,j} B_{j,k}
    $
    where:
    - $ i $: Row index of matrix $ \mathbf{C} $.
    - $ k $: Column index of matrix $ \mathbf{C} $.
    - $ j $: Index of summation, iterating over all elements in the row $ i $ of $ \mathbf{A} $ and column $ k $ of $ \mathbf{B} $.

##### Practical Interpretation:
* **Matrix multiplication** combines two matrices $ \mathbf{A} $ and $ \mathbf{B} $ by aligning rows of $ \mathbf{A} $ with columns of $ \mathbf{B} $, summing the element-wise products.
* Commonly used in:
  - Solving systems of linear equations.
  - Computer graphics transformations.
  - Machine learning and deep learning computations.


```python
A = np.array([[3,4], [5,6], [7,8]])
b = np.array([1,2])

# Auto multiply
result = np.dot(A, b)

# Manual multiply
result_expanded = []

for i in range(len(A)):
    sum_ = 0
    for j in range(len(b)):
        sum_ += A[i][j] * b[j]
    result_expanded.append(sum_)

result_expanded
```

# Matrix Inversion (Regression)

* Clever, convenient approach for solving linear equations
* An alternative to manually solving with substitution or elimination
* Matrix inverse of $X$ is denoted as $X^{-1}$
  * Satisfies: $ X^{-1} X = I_n $ (Identity Matrix)

<img src="images/matrix_inversion.png" style="width:500px; height:auto;">

In the equation $ y = Xw $
* We know the outcomes $y$, witch could be house prices
* We know the features $X$, witch are predictors like bedroom count
* Vector $w$ contains the unknowns, the model's learnable parameters

Assuming $X^{-1}$ exists, matrix inversion can solve for $w$:


$Xw = y$

$X{-1} Xw = X^{-1}y$

$I_n w = X^{-1} y$

$w = X^{-1} y $

<img src="images/matrix_inversion_2.png" style="width:500px; height:auto;">

```python
X = np.array([[4, 2], [-5, -3]])
X
```

```python
Xinv = np.linalg.inv(X)
Xinv
```

```python
y = np.array([4, -7])
y
```

```python
w = np.dot(Xinv, y)
w
```

```python
y = np.dot(X, w)
y
```

# Diagonal Matrices

* Nonzero elements along main diagonal; zeros everywhere else
* Identity Matrix $I_n$ is an example
* If square, denoted as $diag(x)$ where $x$ is vector of main-diagonal elements
* Computationally efficient
    * Multiplication: $diag(x)y = x \circ y $
    * Inversion: $diag(x)^{-1} = diag[1/x,...,1/x]^{t}$
        * Can't divide by zero, so $x$ can't include zero
* Can be no-square and computation still eficient:
    * if $h$ > $w$, simply add zeros to product
    * if $w$ > $h$, remove elements from product

# Unit Vectors

* A normalized vector with a magnitude equal to 1
* Represents only the direction of a vector, independent of its length
* Obtained by dividing a vector by its norm:
  * Expression: $\mathbf{u} = \frac{\mathbf{v}}{\|\mathbf{v}\|}$ (ensures: $\|\mathbf{u}\| = 1$)


```python
v = np.array([78, 39, 20, 15, 8])

norm_v = np.linalg.norm(v)

unit_vector = v / norm_v

# All values should be 1
norm = np.linalg.norm(unit_vector)
norm
```

# Orthogonal Matrices

In orthogonal matrices, orthonormal vectors:
* Make up all rows
* Make up all columns

This means: $A^{T}A = AA^{T} = I$

Which also means: $A^{T} = A^{-1}I = A^{-1}$'

Calculating $A^{T}$ is cheap, therefore so is calculating $A^{-1}$

```python
theta = np.radians(30)
A = np.array([
    [np.cos(theta), -np.sin(theta)],
    [np.sin(theta),  np.cos(theta)]
])

# Verify the orthogonality: A^T * A should equal the identity matrix
identity = np.dot(A.T, A)

print("Rotation Matrix Q:")
print(A)
print("\nQ^T * Q (should be the identity matrix):")
print(identity)
```
